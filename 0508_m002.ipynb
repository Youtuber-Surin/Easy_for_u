```python
# n 차식의 수학 분석

# 차수와 계수
quiz = ["a", "3a", "-7ab", "3a-16ab", "77x+2y"]

for i in range( len(quiz) ) :
    print("문제 :", quiz[i], "에 대한 질문의 답변하시오.")
    
    print("도움말 : 항은 상수/변수들의 곱이 된 식")
    print("도움말 : 차수는 각 항이 변수들의 곱이 된 변수의 수")
    print("도움말 : 계수는 각 항에서 변수들을 제외한 유리수 값")
    
    print(quiz[i], "는 상수/변수들의 곱으로 이뤄진 식이 맞습니다.")
    
    if i > 2 :
        print("다만,", quiz[i], "는 여러 곱의 식으로 이뤄진 다항식이 맞습니다.")
    else :
        print("다만,", quiz[i], "는 여러 곱의 식으로 이뤄진 단항식이 맞습니다.")
    
    if i == 0 :
        print(quiz[i], "의 차수는 1 입니다.")
        print(quiz[i], "의 계수는 1 입니다.")
    elif i == 1 :
        print(quiz[i], "의 차수는 1 입니다.")
        print(quiz[i], "의 계수는 3 입니다.")
    elif i == 2 :
        print(quiz[i], "의 차수는 2 입니다.")
        print(quiz[i], "의 계수는 -7 입니다.")
    elif i == 3 :
        print(quiz[i], "의 차수는 2 입니다.")
        print(quiz[i], "의 계수는 3, -16 입니다.")
    elif i == 4 :
        print(quiz[i], "의 차수는 2 입니다.")
        print(quiz[i], "의 계수는 77, 2 입니다.")
        
    print()
```
